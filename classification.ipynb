```python
def normalize(X):
    means = []
    stds = []
    for name in ("H2", "CO","C2H4","C2H2"):
        m = np.mean(X[name])
        std = np.std(X[name])
        X[name] = (X[name] - m) / std
        print(m)
        means.append(m)
        stds.append(std)
    return X, means, stds
```

```python
res, m, s = normalize(X)
```

```python
print(res, m, s)
```
